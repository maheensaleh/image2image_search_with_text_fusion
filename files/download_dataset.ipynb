```python
# import fiftyone as fo
import fiftyone.zoo as foz


```

```python

# To download the COCO dataset for only the "person" and "car" classes
dataset = foz.load_zoo_dataset(
    "coco-2017",
    splits="train",
    label_types="detections",
    # classes=["person", "cat","book","cup","chair","clock","bench"],
    max_samples=1000
)

# Export the dataset
export_dir = "/content/drive/MyDrive/UdS-DSAI/HLCV/project/dataset3/train"


dataset.export(
    export_dir=export_dir,
    dataset_type=fo.types.COCODetectionDataset,
)
```

```python

# To download the COCO dataset for only the "person" and "car" classes
dataset = foz.load_zoo_dataset(
    "coco-2017",
    splits="validation",
    label_types="detections",
    # classes=["person", "cat"],
    max_samples=100
)

# Export the dataset
export_dir = "/content/drive/MyDrive/UdS-DSAI/HLCV/project/dataset3/valid"


dataset.export(
    export_dir=export_dir,
    dataset_type=fo.types.COCODetectionDataset,
)
```
